```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the CSV file again after code execution reset
csv_path = r"C:\Users\Hitesh\OneDrive\Desktop\padm\Assignment2(version3)\logs\log(8).csv"
df = pd.read_csv(csv_path)

# Convert 'steps' column to numeric (remove braces if present)
df['steps'] = df['steps'].astype(str).str.extract(r'(\d+)').astype(float)

# Filter to only show success/fail streaks
df['Success'] = df['Result'] == 'Success'

# Visualize episode performance
plt.figure(figsize=(12, 5))
sns.lineplot(x=df['Episode'], y=df['steps'], hue=df['Success'], palette={True: 'green', False: 'red'})
plt.axhline(1000, color='gray', linestyle='--', label='Max Steps Threshold')
plt.title('Steps per Episode (Green: Success, Red: Fail)')
plt.xlabel('Episode')
plt.ylabel('Steps')
plt.legend()
plt.tight_layout()
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load the CSV file again after code execution reset
csv_path = r"C:\Users\Hitesh\OneDrive\Desktop\padm\Assignment2(version3)\logs\log(8).csv"
df = pd.read_csv(csv_path)

# Ensure correct column names (case-sensitive)
df.columns = [col.strip() for col in df.columns]  # remove extra spaces

# Convert Episode and Epsilon to numeric
df['Episode'] = df['Episode'].astype(float)
df['epsilon'] = pd.to_numeric(df['epsilon'], errors='coerce')

# Plot Epsilon Decay
plt.figure(figsize=(12, 5))
sns.lineplot(x='Episode', y='epsilon', data=df, color='blue')
plt.title('Epsilon Decay over Episodes')
plt.xlabel('Episode')
plt.ylabel('Epsilon')
plt.grid(True)
plt.tight_layout()
plt.savefig(r"D:\Thi\Padm\epsilondecay_over_epoch")
plt.show()


```
